```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,100)
```

```python
x[0]
```

```python
def plotcos(f):
    plt.plot(x, np.cos(2*math.pi * f *x))
```

```python
plotcos(1)
```

```python
plotcos(1.2)
```

```python
import ipywidgets
```

```python
f_control = ipywidgets.FloatSlider(value = 1, min = 0.5, max = 2.0, step = 0.05, continuous_update = False)
```

```python
ipywidgets.interact(plotcos, f = f_control);
```

```python
x = np.linspace(-2*np.pi, 2*np.pi, 1000)
def plotsin(f):
    plt.plot(x, f*np.sin(x))
```

```python
f_control = ipywidgets.FloatSlider(value = 1, min = 0, max = 8, step = 1, continuous_update = False)
```

```python
ipywidgets.interact(plotsin, f = f_control);
```

```python
def random_walk(numsteps):
    stepsize = 1
    position = []
    x = 0
    for i in range(int(numsteps)):
        number = np.random.rand()
        if number < 0.5:
            x = x - stepsize
        else:
            x = x + stepsize
        position = position + [x]
    plt.plot(position, marker= '.', linestyle = '')
```

```python
f_control = ipywidgets.FloatSlider(value = 100, min = 10, max = 1000, step = 10, continuous_update = False)
```

```python
ipywidgets.interact(random_walk, numsteps = f_control);
```
